Data Wrangling
- 

```python
unique = []
for molecule in train_df["SMILES"]:
    for element in molecule:
        if element not in unique:
            unique.append(element)
```

```python
unique = [x for x in unique if x not in '123456789(=)[@]+-\\#/'] #this value can be changed for further using of single bonds, double bonds, etc. 
```

```python
unique_df = pd.DataFrame(columns = unique)
for molecule in train_df["SMILES"]:
    my_dict = {key: 0 for key in unique}   
    for element in molecule:
        if element in unique:
            my_dict[element] = my_dict[element] + 1
    unique_df.loc[len(unique_df)] = my_dict

train_df = train_df.drop("SMILES", axis=1)
train_df = pd.concat([train_df, unique_df], axis=1)
train_df
```

```python
train_df['Label'] = train_df['Label'].map({'Yes': 1, 'No': 0})
train_df
```

```python
unique_df = pd.DataFrame(columns = unique)
for molecule in test_df["SMILES"]:
    my_dict = {key: 0 for key in unique}   
    for element in molecule:
        if element in unique:
            my_dict[element] = my_dict[element] + 1
    unique_df.loc[len(unique_df)] = my_dict

test_df = test_df.drop("SMILES", axis=1)
test_df = pd.concat([test_df, unique_df], axis=1)
test_df
```

```python
test_df["Label"] = test_df["Label"].map({"Yes": 1, "No": 0})
test_df["Label"]
```

```python
from sklearn.preprocessing import StandardScaler
#subtract by mean then divide by std devf

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(train_df.drop("Label", axis=1))
X_train_scaled
```

```python
X_test_scaled = scaler.fit_transform(test_df.drop("Label", axis=1))
X_test_scaled
```

```python
X_scaled = pd.concat([pd.DataFrame(X_train_scaled), pd.DataFrame(X_test_scaled)])
pd.DataFrame(X_train_scaled)
X_scaled
```

```python
X_train = train_df.drop("Label", axis=1)
y_train = train_df["Label"]
X_test = test_df.drop("Label", axis=1)
y_test = test_df["Label"]
```

```python
y = pd.concat([y_train, y_test])
y
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.DataFrame({
    'Model': ['LinearRegression', 'Ridge', 'Lasso', 'DecisionTree', 'RandomForest', 'SVR', 'KNN', 'XGBoost'],
    'R2 Score': [0.85, 0.83, 0.81, -0.12, 0.88, -0.03, 0.76, 0.91]
})

bar_colors = ['green' if val >= 0 else 'red' for val in df['R2 Score']]
sns.set_style("whitegrid")
plt.figure(figsize=(14, 7))

bars = sns.barplot(x='R2 Score', y='Model', data=df, palette=bar_colors)

for i, (score, model) in enumerate(zip(df['R2 Score'], df['Model'])):
    bars.text(score + (0.01 if score >= 0 else -0.05), i, f"{score:.2f}", 
              color='black', va='center', fontweight='bold')

plt.title('Model Performance Comparison (RÂ² Score)', fontsize=16, weight='bold')
plt.xlabel('RÂ² Score')
plt.ylabel('ML Models')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.DataFrame({
    'Model': ['LinearRegression', 'Ridge', 'Lasso', 'DecisionTree', 'RandomForest', 'SVR', 'KNN', 'XGBoost'],
    'RMSE': [0.42, 0.51, 0.52, 0.53, 0.36, 0.46, 0.43, 0.34]
})

rmse_median = df['RMSE'].median()
bar_colors = ['green' if val <= rmse_median else 'red' for val in df['RMSE']]

sns.set_style("whitegrid")
plt.figure(figsize=(14, 7))

bars = sns.barplot(x='RMSE', y='Model', data=df, palette=bar_colors)

for i, (score, model) in enumerate(zip(df['RMSE'], df['Model'])):
    bars.text(score + 0.01, i, f"{score:.2f}", color='black', va='center', fontweight='bold')

plt.title('Model Performance Comparison (RMSE)', fontsize=16, weight='bold')
plt.xlabel('Root Mean Squared Error (RMSE)')
plt.ylabel('ML Models')

plt.tight_layout()
plt.show()
```
